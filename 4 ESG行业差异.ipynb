# ESG行业差异

- 创新点：研究ESG评级与股票波动率溢出效应在不同行业和地区之间的差异。可以分析高ESG评级的行业（如可再生能源、科技）与低ESG评级的行业（如传统能源、烟草）之间的波动率溢出效应。
- 研究方法：利用行业分类指数，并结合不同地区（如发达国家与发展中国家、不同地理区域）的数据，进行分组分析，评估ESG评级对这些群体的不同影响。

## 不同行业的ESG平均数值

```python
import pandas as pd
path='datasets/华证2009-2023年（含细分项+季度)）/华证esg评级2009-2023（细分项）/华证esg评级含细分项（年度）2009-2023.xlsx'
ESG_df = pd.read_excel(path, dtype={'股票代码':str})
ESG_df['股票代码'] = ESG_df['股票代码'].str.zfill(6)
ESG_df.head(1)
```

```python
ESG_df.columns[]
```

```python
sorted_ESG = ESG_df[['行业名称', '综合得分']].groupby('行业名称').mean().sort_values('综合得分')
```

```python
ESG_df['行业名称'].unique()
```

```python
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('svg') 
plt.rcParams['font.sans-serif'] = 'Songti SC'
plt.rcParams['axes.unicode_minus'] = False  # 解决负号“-”显示异常
```
